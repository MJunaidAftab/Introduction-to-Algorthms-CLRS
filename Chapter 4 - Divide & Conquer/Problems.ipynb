$$

= z + z\sum\limits_{i=2}^{\infty} F_{i-1}z^{i-1} + z^2\sum\limits_{i=2}^{\infty} F_{i-2}z^{i-2}

b. We have

$$
F(z)
=
\frac{z}{1-z-z^2}
$$

Factoring the denominator with the quadratic formula shows $1 − z − z^2 = (1 − \varphi z)(1 − \widehat{\varphi}z )$, and the final equality comes from a partial fraction decomposition.

c.

$$F(z) = \frac{1}{\sqrt{5}} \left(\frac{1}{1-\phi z} - \frac{1}{1-\hat{\phi}z}\right) = \frac{1}{\sqrt{5}} \sum\limits_{i=0}^{\infty}(\phi z)^i - \frac{1}{\sqrt{5}} \sum\limits_{i=0}^{\infty}(\hat{\phi}z)^i = \sum\limits_{i=0}^{\infty}\frac{1}{\sqrt{5}}(\phi^i - \hat{\phi}^i)z^i$$

d.

From the definition of the generating function, $F_i$ is the coefficient of $z^i$ in $F(z)$. By part (c) this is given by $\frac{1}{\sqrt{5}}(\phi^i - \hat{\phi}^i)$. Since $|\hat{\phi}| < 1$ we must have $\left|\frac{\hat{\phi}^i}{\sqrt{5}}\right| < \left|\frac{\hat{\phi}}{\sqrt{5}}\right| < \frac{1}{2}$. Finally, since the Fibonacci numbers are integers we see that the exact solution must be the approximated solution $\frac{\phi^i}{\sqrt{5}}$ rounded to the nearest integer.

# Problem 6

a. 

If an array $A$ is Monge then trivially it must satisfy the inequality by taking
$k = i + 1$ and $l = j + 1$. Now suppose $A[i, j] + A[i + 1, j + 1] \leq A[i, j + 1] +
A[i + 1, j]$. We'll use induction on rows and columns to show that the array $A$ satisfies the Monge property.
Now fix $i$ and $j$, let $r \geq 1$, and suppose that $A[i, j]+A[i+1, j+r] \leq A[i, j+r]+A[i+1, j]$.
By applying the induction hypothesis and the given inequality, we have:

$$
A[i, j] + A[i + 1, j + r + 1] \leq A[i, j + r] + A[i + 1, j] - A[i + 1, j + r] + A[i, j + r + 1] + A[i + 1, j + r] - A[i, j + r] = A[i + 1, j] + A[i, j + r + 1]
$$

so it follows that we can extend columns and preserve the Monge property.
Next we induct on rows. 
Suppose that $A[i, j] + A[k, l] ≤ A[i, l] + A[k, j]$.
Then we have

\begin{align*}
A[i, j] + A[k + 1, l] & \leq A[i, l] + A[k, j] - A[k, l] + A[k + 1, l]  \\
& \leq A[i, l] + A[k, j] - A[k, l] + A[k, l] + A[k + 1, l - 1] - A[k, l - 1] \\
& = A[i, l] + (A[k, j] + A[k + 1, l - 1]) - A[k, l - 1] \\
& \leq A[i, l] + A[k, l - 1] + A[k + 1, j] - A[k, l - 1] \\
& = A[i, l] + A[k + 1, j].
\end{align*}

b. Change the 7 to a 5.

c. Suppose that there exist $i$ and $k$ such that $i < k$ but $f(i) > f(k)$.  Since $A$ is Monge we must have
$A[i, f(k)] + A[k, f(i)] \leq A[k, f(k)] + A[i, f(i)]$.

Since $f(i)$ gives the position of the leftmost minimum element in row $i$, this implies that $A[i, f(i)] < A[i, f(k)]$. 
Moreover, $A[k, f(k)] \leq A[k, f(i)]$. 
Combining these we get $A[k, f(k)] + A[i,f(i)] \leq A[k,f(i)] + A[i,f(k)]$.
This contradicts the Monge array property.

d. Based on the conclusion of part c we can restrict the elements which we need to examine. For each odd row $k$ we need only consult the elements between $f(k-1)$ and $f(k+1)$, where we take $f(0) = 1$ and $f(m) = f(m+1) = n$.
The matrix has $\lceil m/2 \rceil$ odd-numbered rows, and so finding the leftmost minimum of each of them takes:

\begin{equation*}
\sum_{k=1}^{\lceil m/2 \rceil} (f(k+1) - f(k-1) + 1) 
= \lceil m/2 \rceil + \sum_{k=1}^{\lceil m/2 \rceil} (f(k+1) - f(k-1))
= O(m) + f(m)/f(m+1)
= O(m+n)
\end{equation*}
