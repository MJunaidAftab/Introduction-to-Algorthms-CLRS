## Problem 3

We exploit the fact that $\mathbf{P} ( 01 ) = \mathbf{P} ( 10 )$. We call the UNBIASED-RANDOM function twice. If both bits are different, we output $0$ if the bits are $0$ and $1$ (in order) and we we output $1$ if the bits are $1$ and $0$ (in order). If both bits are the same, we call repeat the above procedure.

Here is the pseudocode:

```python
def unbiased_random(p):
        
    Boolean = True
    
    while Boolean == True:
    
        x = biased_random()
        y = biased_random()
        
        if x != y:
            
            if x == 0:
                
                return x
            
            else:
                
                return y
            
            Boolean = False
```

The expected run time of thr algorithm is $ \Theta ( 1 / 2p(1-p) )$ since the decision to repeat or not repeat the random number generation process can be modelled as a sequence of Bernoulli trials, each with success probability $ 2p(1-p) $.

## Problem 1

We hire a candidate exactly 1 time if is first candidate is best among all the candidates. Hence the desired probability is given by,

$$
\prod_{i = 2}^{n} ( 1 - 1/i ) = \frac{i - 1}{i} = \frac{1}{2} \frac{2}{3} \cdots \frac{n-1}{n} = 1/n.
$$

We hire exactly $n$ times if each successive candidate is hired. The probability that candidate $i$ is hired is $1/i$, so the desired probability is,

$$
\prod_{i = 1}^{n} = \frac{1}{i} = \frac{1}{n!}.
$$

## Problem 2

We can hire exactly twice in a number of ways. Note that we always hire the first candidate, so then any of the second, third and so forth candidates can be hired. Let $p_{1, i}$ denote the probability that candidates 1 and $i, i = 2, \cdots n$ are hired. This probability is given by,

$$
p_{1, i} = 1/i \prod_{j = 2, j \neq i}^{n} ( 1 - 1/j ) = 1/i \prod_{j = 2, j \neq i}^{n} \frac{j - 1}{j} = \frac{1}{2} \frac{2}{3} \cdots \frac{i - 2}{i - 1} \frac{1}{i} \frac{i}{i + 1} \cdots \frac{n - 1}{n} = \frac{1}{i n}.
$$

Since the events are mutually exclusive, the desired property is given by,

$$
\sum_{i = 2}^{n} p_{1, i} = \frac{1}{i n} = \frac{ \sum_{i = 2}^{n} 1/i}{n} = O \bigg ( \frac{\lg n }{n} \bigg ).
$$

```python
def random_search(A, x):
    
    n = len(A)
    B = [False]*n
    
    count = 0
    
    while count <= n-1:
        
        i = random.randint(0,n-1)
        
        if A[i] == x:
            
            return i
        
        if B[i] != False:
            
            count = count + 1
            B[i] = i
            
            print(B)
            
    return 'The list does not contain the element.'
```
